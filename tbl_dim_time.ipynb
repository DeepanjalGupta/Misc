```python
import numpy as np
import pandas as pd
import datetime as dt
from dateutil.relativedelta import relativedelta
```

```python
#---------------------------------------------------------------------
#---------------------------------------------------------------------
date_label_list = []
d1 = dt.date(2000, 1, 1)    # start date
d2 = dt.date(2050, 12, 31)  # end date

delta = d2 - d1             # timedelta

for i in range(delta.days + 1):
    date_label_list.append((d1 + dt.timedelta(i)).strftime("%Y-%m-%d"))
#---------------------------------------------------------------------
#---------------------------------------------------------------------    
date_id_list = []
for i in range(1,len(date_label_list)+1):
    date_id_list.append(i)   
#---------------------------------------------------------------------
#---------------------------------------------------------------------
date_type_id_list = []
for i in range(1,len(date_label_list)+1):
    date_type_id_list.append(1)  
#---------------------------------------------------------------------
#---------------------------------------------------------------------    
date_type_list = []
for i in range(1,len(date_label_list)+1):
    date_type_list.append('DAY')    
#---------------------------------------------------------------------
#---------------------------------------------------------------------
```

```python
df_tbl_dim_time_day = a = pd.DataFrame({'date_id':date_id_list
                                        ,'date_label':date_label_list
                                        ,'date_type_id':date_type_id_list
                                        ,'date_type_label':date_type_list})
```
